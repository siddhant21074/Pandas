# Insert

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4,5],"B":[5,6,7,8,9]})
var

var.insert(2,"C",[10,11,12,13,14])   # Here we inserted new the data using list into C var
var

# We can copy the data of one variable into another like 

var.insert(3,"D",var["C"])  # Here we inserted the data of c into the variable D on 3rd location
var
```

```python
# Also we can add the required data we want

var["E"] = var["A"][:3]   # Here we created the new column E and added the data of A from 0 to index no 3
var
```
