# melt Function()

```python
import pandas as pd

var = pd.DataFrame({"Name":["a","b","c","d","e"],"Date":[1,2,3,4,5],"Roll No.":[10,20,30,40,50]})
pd.melt(var)

# This function is used to arrange the element in the dataframe in  column wise manner i.e name-date and then rollno

```

```python
var1 = pd.melt(var,id_vars = "Date")
var1

# Here we can assign any data as a index in the dataframe such as date we have assigned
```

```python
var2 = pd.melt(var,var_name = "Title",value_name = "Number")
var2
# Here we can change name of the variable as shown above 
# Same applies to value
```

# pivot Function()

```python
import pandas as pd

n = pd.DataFrame({"English":[10,11,15,18],"History":[14,12,15,19],"Division":["A","B","A","B"],"Date":[1,2,3,4]})
var6 = n.pivot(index = "Date",columns = "Division")
var6
```

```python
var7 = n.pivot_table(index = "Date",columns = "Division",aggfunc = "mean")
var7


# NOTE : to perfoem certain arithmetic operation we have to change the .pivot to .pivot_table 
```

```python
var8 = n.pivot_table(index = "Date",columns = "Division",aggfunc = "sum")
var8
# Here we can find the sum the mean and etc
```

```python
var9 = n.pivot_table(index = "Date",columns = "Division",margins = "True")
var9
# Here we get an average value from the margin and stored into the ALL column 
```
